This is the source of the `Vector` class; you will need to modify it for exercises 9.A and 9.B.

```python
class Vector(object):
  """A 2D Vector."""
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def add(self, other):
    return Vector(self.x+other.x,
                  self.y+other.y)
  def mul(self, scalar):
    return Vector(scalar*self.x, scalar*self.y)
  def show(self):
    return ("<%g,%g>" % (self.x, self.y))
```

```python
from math import sqrt

class Vector(object):
  """A 2D Vector."""
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def add(self, other):
    return Vector(self.x+other.x,
                  self.y+other.y)
  def mul(self, scalar):
    return Vector(scalar*self.x, scalar*self.y)
  def show(self):
    return ("<%g,%g>" % (self.x, self.y))
  def norm(self):  # <-- THIS IS NEW
    return sqrt(self.x**2 + self.y**2)
```

## Exercise 9.B

Add a new method `unit` to the `Vector` class: 
if `v` is an instance of class `Vector`, then
calling `v.unit()` returns the vector `u` having the same
direction as `v` but norm 1.

```python
# build upon modified Vector of Exercise 9.A

from math import sqrt

class Vector(object):
  """A 2D Vector."""
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def add(self, other):
    return Vector(self.x+other.x,
                  self.y+other.y)
  def mul(self, scalar):
    return Vector(scalar*self.x, scalar*self.y)
  def show(self):
    return ("<%g,%g>" % (self.x, self.y))
  def norm(self):
    return sqrt(self.x**2 + self.y**2)
  def unit(self):  # <-- THIS IS NEW
    return self.mul(1 / self.norm())
```

If your solution is correct, applying the `.unit()` method to `a = Vector(3, 0)` and `b = Vector(0, 4)` should yield the basis vectors `<1, 0>` and `<0, 1>` as result.

```python
a = Vector(3, 0)
b = Vector(0, 4)

e1 = a.unit()
e2 = b.unit()

e1.show(), e2.show()
```
